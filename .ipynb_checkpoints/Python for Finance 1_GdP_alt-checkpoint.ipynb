```python
import pandas as pd
import datetime as dt
from pandas_datareader import data as web
import mplfinance as mpf

BASE_DIR = r"C:\Users\gerha\Github\Python4Finance"

def _csv_path(ticker: str) -> str:
    t = ticker.strip().lower()
    return f"{BASE_DIR}\\Python4Finance{t}.csv"

# 1) Download from Stooq and save with Date as index (sorted ascending)
def save_to_csv_from_stooq(ticker, syear, smonth, sday, eyear, emonth, eday):
    start = dt.datetime(syear, smonth, sday)
    end   = dt.datetime(eyear, emonth, eday)

    # Stooq returns a DatetimeIndex already; often newest->oldest, so sort it.
    df = web.DataReader(ticker, "stooq", start, end).sort_index()

    # Ensure the index name is Date (nice for readability)
    df.index.name = "Date"

    # Save
    path = _csv_path(ticker)
    df.to_csv(path)

    return df

# 2) Read CSV back (Date as index) â€“ helper (optional but keeps things consistent)
def get_df_from_csv(ticker):
    path = _csv_path(ticker)
    try:
        df = pd.read_csv(path, index_col=0, parse_dates=True)
    except FileNotFoundError:
        print("File Doesn't Exist:", path)
        return None

    # Safety: ensure sorted and clean index
    df.index = pd.to_datetime(df.index, errors="coerce")
    df = df[~df.index.isna()].sort_index()
    df.index.name = "Date"
    return df

# 3) Add daily return and overwrite CSV (still Date as index)
def add_daily_return_to_df(ticker):
    df = get_df_from_csv(ticker)
    if df is None:
        return None

    # daily_return based on Close
    df["daily_return"] = df["Close"].pct_change()

    # Save back
    path = _csv_path(ticker)
    df.to_csv(path)
    return df

# 4) Plot using mplfinance over a date range
def mplfinance_plot(ticker, chart_type, syear, smonth, sday, eyear, emonth, eday):
    start = pd.Timestamp(syear, smonth, sday)
    end   = pd.Timestamp(eyear, emonth, eday)

    df = get_df_from_csv(ticker)
    if df is None:
        return

    # Slice works because index is monotonic (sorted)
    df_sub = df.loc[start:end]

    if df_sub.empty:
        print(f"No rows found for {ticker} between {start.date()} and {end.date()}")
        return

    # mplfinance expects at least Open/High/Low/Close (and Volume if volume=True)
    mpf.plot(
        df_sub,
        type=chart_type,
        mav=(3, 5, 7),
        volume=("Volume" in df_sub.columns),
        show_nontrading=True
    )
```

```python
save_to_csv_from_stooq("amzn", 2020, 1, 1, 2021, 1, 1)
add_daily_return_to_df("amzn")
mplfinance_plot("amzn", "ohlc", 2020, 6, 1, 2021, 1, 1)
```
